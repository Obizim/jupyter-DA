#### Cleaning the data: removing the dollar sign and the commas in the figures 
#### Also, converting the data types of the transformed figures to intergers

```python
def transform_figures(text): 
    text = text.str.replace('$', '').str.replace(',', '')
    text = pd.to_numeric(text)
    return text
```

```python

df['Worldwide Box Office'] = transform_figures(df['Worldwide Box Office'])
df['Domestic Box Office'] = transform_figures(df['Domestic Box Office'])
df['Opening Weekend'] = transform_figures(df['Opening Weekend'])
df['Production Budget'] = transform_figures(df['Production Budget'])
```

```python
roi = df['Worldwide Box Office'] - df['Production Budget']
df['ROI'] = roi
```

#### Percentages of domestic box office and worldwide box office 

```python
total_domestic_box_office = df['Domestic Box Office'].sum()
total_worldwide_box_office = df['Domestic Box Office'].sum()
```

```python
df.drop('% domestic box office', axis=1, inplace=True)
```

```python
df['Domestic Box Office Percentage'] = (df['Domestic Box Office'] / total_domestic_box_office) * 100
df['Worldwide Box Office Percentage'] = (df['Worldwide Box Office'] / total_domestic_box_office) * 100
```

```python
df
```

```python
chart = [df['Domestic Box Office Percentage'].sum(), df['Worldwide Box Office Percentage'].sum()]
labels = ['Domestic Box Office Percentage', 'Worldwide Box Office Percentage']
plt.pie(chart , labels = labels, autopct='%1.1f%%', radius=0.9)
plt.show()
```
