#### Cleaning the data: removing the dollar sign and the commas in the figures 
#### Also, converting the data types of the transformed figures to intergers

```python
def transform_figures(text): 
    text = text.str.replace('$', '').str.replace(',', '')
    text = pd.to_numeric(text)
    return text
```

```python

df['Worldwide Box Office'] = transform_figures(df['Worldwide Box Office'])
df['Domestic Box Office'] = transform_figures(df['Domestic Box Office'])
df['Opening Weekend'] = transform_figures(df['Opening Weekend'])
df['Production Budget'] = transform_figures(df['Production Budget'])
```

```python
roi = df['Worldwide Box Office'] - df['Production Budget']
df['ROI'] = roi
```
